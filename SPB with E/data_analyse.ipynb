```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
import scipy.constants as cst
from labellines import labelLines
import pickle
import datetime
import pandas as pd
import statsmodels.api as sm
plt.rcParams.update({'font.size': 14})
import pickle
```

```python
for_lowE_data_file = open('forbidden_smallE_photons.pickle', 'rb')
for_data_file = open('forbidden_photons.pickle', 'rb')
for_lowE_data = pickle.load(for_lowE_data_file)
for_data = pickle.load(for_data_file)
```

```python
Ez_reg = np.linspace(1,200,11)
pow_reg = np.linspace(0.01,1,20)
```

## Plot Forbidden transition photons for low E

```python
fig_fe, ax_fe = plt.subplots(figsize = (8,5)) 
# for i in range(len(Ez_reg)):
#     ax_fe.plot(pow_reg, for_lowE_data[i], '.-',label = f"{Ez_reg[i]}V/cm")

ax_fe.plot(pow_reg, for_lowE_data[-1], '.-',label = f"{Ez_reg[-1]}V/cm")
ax_fe.plot(pow_reg, for_lowE_data[-2], '.-',label = f"{Ez_reg[-2]}V/cm")
ax_fe.legend(loc='upper center', bbox_to_anchor=(1.15, 1.),
          ncol=1, fancybox=True, shadow=True)
ax_fe.set_xlabel('laser power[W]')
ax_fe.set_ylabel('# of photons')
ax_fe.set_title('# of photons from "forbidden" transition')
```

## Plot Forbidden transition photons for all E

```python
fig_fE, ax_fE = plt.subplots(figsize = (8,5)) 
# for i in range(len(Ez_reg)):
#     ax_fe.plot(pow_reg, for_lowE_data[i], '.-',label = f"{Ez_reg[i]}V/cm")

ax_fE.plot(pow_reg, for_data[-1], '.-',label = f"{Ez_reg[-1]}V/cm")
ax_fE.plot(pow_reg, for_data[-2], '.-',label = f"{Ez_reg[-2]}V/cm")
ax_fE.legend(loc='upper center', bbox_to_anchor=(1.15, 1.),
          ncol=1, fancybox=True, shadow=True)
ax_fE.set_xlabel('laser power[W]')
ax_fE.set_ylabel('# of photons')
ax_fE.set_title('# of photons from "forbidden" transition')
```
